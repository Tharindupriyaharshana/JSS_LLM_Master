```python
# ============================================================
# 🧠 1. Imports and setup
# ============================================================
from langchain_openai import ChatOpenAI, OpenAIEmbeddings
from langchain_community.vectorstores import Chroma
from langchain.prompts import ChatPromptTemplate
import os, time, pandas as pd, numpy as np

# --- Compatibility patches (NumPy + Chroma) -----------------
if not hasattr(np, "float_"): np.float_ = np.float64
if not hasattr(np, "int_"): np.int_ = np.int64
if not hasattr(np, "bool_"): np.bool_ = np.bool_

import chromadb
if not hasattr(chromadb, "config"):
    import types
    chromadb.config = types.SimpleNamespace(Settings=chromadb.Settings)

# --- 🔑 Enter your API key ---
os.environ["OPENAI_API_KEY"] = "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"


# ============================================================
# 🧩 2. Build vector database for Tool Lifetime Knowledge
# ============================================================
os.makedirs("vector_db/tool_lifetime", exist_ok=True)
os.makedirs("data", exist_ok=True)

tool_lifetime_knowledge = """
Tool Lifetime Knowledge Base
----------------------------

1️⃣ Every tool has a defined lifetime limit (e.g., T1 = 100 minutes).
2️⃣ Tool lifetime consumption is cumulative across all operations using that tool set.
3️⃣ When total usage of a tool (sum of durations in its set) exceeds its lifetime, a replacement is required.
4️⃣ Replacement time is logged at the moment cumulative duration surpasses the lifetime.
5️⃣ Shared tools across machines share lifetime usage; replacements can occur independently per machine.
6️⃣ Goal: predict replacement events precisely by tracking cumulative time across shared sets.

Example:
If T1 has 100 min lifetime and operates for 320 min on M1, replacements occur at 100, 200, and 300 min.
"""

with open("data/tool_lifetime_rules.txt", "w") as f:
    f.write(tool_lifetime_knowledge)

embeddings = OpenAIEmbeddings(model="text-embedding-3-small")
rag_db = Chroma.from_texts(
    texts=[open("data/tool_lifetime_rules.txt").read()],
    embedding=embeddings,
    persist_directory="./vector_db/tool_lifetime",
)
rag_db.persist()

# ============================================================
# 🤖 3. LLM setup
# ============================================================
llm = ChatOpenAI(model="gpt-4o-mini", temperature=0)

# ============================================================
# 💬 4. Prompt templates
# ============================================================

# --- Baseline Prompt ---
baseline_template = ChatPromptTemplate.from_template(
    """You are a simple tool monitoring assistant.
From the schedule, tool usage, and lifetime, identify when each tool might need replacement.
Provide a short summary.

Schedule:
{schedule}

Tool Usage:
{tool_usage}

Tool Sharing Plan:
{sharing_plan}

Tool Lifetime:
{tool_lifetime}
"""
)

# --- Detailed Prompt ---
detailed_template = ChatPromptTemplate.from_template(
    """You are an expert in Job Shop Scheduling (JSS) with integrated tool lifetime tracking.
Given:
1. A Job Shop Schedule (start, duration, end, machine)
2. A Tool Requirement List (which tool each operation uses)
3. A Tool Sharing Plan (how tools are shared)
4. Tool Lifetime Limits (e.g., T1 = 100 min)

Determine when and where tool replacements are required
based on cumulative usage across shared operations.

Output Format:
Tool Replacement Timeline
Tool | Machine | Replacement Time | Reason

Schedule:
{schedule}

Tool Usage:
{tool_usage}

Tool Sharing Plan:
{sharing_plan}

Tool Lifetime:
{tool_lifetime}
"""
)

# --- RAG + Context Prompt ---
rag_template = ChatPromptTemplate.from_template(
    """You are an expert in Job Shop Scheduling with integrated tool lifetime management.
Use this reference knowledge:
{context}

Given:
- Schedule
- Tool usage list
- Tool sharing plan
- Lifetime limits

Predict when each tool requires replacement
based on cumulative usage across shared sets.

Follow this format:
Tool Replacement Timeline
Tool | Machine | Replacement Time | Reason

Schedule:
{schedule}

Tool Usage:
{tool_usage}

Tool Sharing Plan:
{sharing_plan}

Tool Lifetime:
{tool_lifetime}
"""
)

# ============================================================
# 🧮 5. Run evaluation pipeline (2 samples only for testing)
# ============================================================

df = pd.read_excel("/content/valid_50_with_tool_lifetime (1).xlsx")

# Make sure you add columns: Tool_Sharing_Plan, Tool_Lifetime
results = []

for i, row in enumerate(df.head(2).iterrows(), start=1):  # limit to 2 samples
    _, row = row
    qid = row["ID"]
    schedule = row["Question"]
    tool_usage = row["Tool_Assignment"]
    sharing_plan = row.get("Tool_Sharing_Plan", "No sharing plan provided.")
    tool_lifetime = row.get("Tool_Lifetime", "T1=100, T2=150, T3=200")

    print(f"🔧 Processing {qid}...")

    # --- Baseline ---
    base_msgs = baseline_template.format_messages(
        schedule=schedule,
        tool_usage=tool_usage,
        sharing_plan=sharing_plan,
        tool_lifetime=tool_lifetime,
    )
    base_out = llm.invoke(base_msgs).content

    # --- Detailed ---
    det_msgs = detailed_template.format_messages(
        schedule=schedule,
        tool_usage=tool_usage,
        sharing_plan=sharing_plan,
        tool_lifetime=tool_lifetime,
    )
    det_out = llm.invoke(det_msgs).content

    # --- RAG + Context ---
    docs = rag_db.similarity_search("tool lifetime rules", k=2)
    context = "\n".join([d.page_content for d in docs])
    rag_msgs = rag_template.format_messages(
        context=context,
        schedule=schedule,
        tool_usage=tool_usage,
        sharing_plan=sharing_plan,
        tool_lifetime=tool_lifetime,
    )
    rag_out = llm.invoke(rag_msgs).content

    print(f"✅ Completed {qid}")

    results.append(
        {
            "ID": qid,
            "Schedule": schedule,
            "Tool_Usage": tool_usage,
            "Tool_Sharing_Plan": sharing_plan,
            "Tool_Lifetime": tool_lifetime,
            "Baseline_Output": base_out.strip(),
            "Detailed_Output": det_out.strip(),
            "RAG_Output": rag_out.strip(),
        }
    )

    time.sleep(0.5)

# ============================================================
# 💾 6. Save final results
# ============================================================
os.makedirs("outputs", exist_ok=True)
out_path = "outputs/tool_lifetime_comparison.xlsx"
pd.DataFrame(results).to_excel(out_path, index=False)
print(f"\n✅ Results saved to {out_path}")
```
