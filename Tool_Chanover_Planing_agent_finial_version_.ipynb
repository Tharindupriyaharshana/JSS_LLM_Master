```python
# ============================================================
#  1. Imports and setup
# ============================================================
from langchain_openai import ChatOpenAI, OpenAIEmbeddings
from langchain_community.vectorstores import Chroma
from langchain.prompts import ChatPromptTemplate
import os, time, pandas as pd, numpy as np

# --- Compatibility patches (NumPy + Chroma) -----------------
if not hasattr(np, "float_"): np.float_ = np.float64
if not hasattr(np, "int_"): np.int_ = np.int64
if not hasattr(np, "bool_"): np.bool_ = np.bool_

import chromadb
if not hasattr(chromadb, "config"):
    import types
    chromadb.config = types.SimpleNamespace(Settings=chromadb.Settings)

# ---  Enter your API key ---
os.environ["OPENAI_API_KEY"] = "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"



# ============================================================
#  2. Build simple vector database for Tool Sharing Knowledge
# ============================================================
os.makedirs("vector_db/tool_sharing", exist_ok=True)
os.makedirs("data", exist_ok=True)

tool_knowledge = """
Tool Sharing Knowledge Base
----------------------------
1 Tools are shared across operations but cannot be used by two machines simultaneously.
2 When two operations require the same tool at overlapping times, multiple tool sets are needed.
3 When a tool finishes on one machine and starts on another after a gap, it can be shared (same set).
4 Each tool type (T1T5) must have its own allocation plan.
5 The goal is to minimize the number of tool sets while avoiding overlap conflicts.
"""

with open("data/tool_rules.txt", "w") as f:
    f.write(tool_knowledge)

embeddings = OpenAIEmbeddings(model="text-embedding-3-small")
rag_db = Chroma.from_texts(
    texts=[open("data/tool_rules.txt").read()],
    embedding=embeddings,
    persist_directory="./vector_db/tool_sharing",
)
rag_db.persist()

# ============================================================
#  3. LLM setup
# ============================================================
llm = ChatOpenAI(model="gpt-4o-mini", temperature=0)

# ============================================================
#  4. Prompt templates
# ============================================================

# --- Baseline / Simple Prompt ---
baseline_template = ChatPromptTemplate.from_template(
    """You are a scheduling assistant.
From the given schedule and tool usage, create a simple summary of tool sharing needs.
Only describe which tools are required and if multiple sets are likely needed.

Schedule:
{schedule}

Tool Usage:
{tool_usage}
"""
)

# --- Detailed Prompt ---
detailed_template = ChatPromptTemplate.from_template(
    """You are an expert in Job Shop Scheduling tool allocation.
Analyze tool sharing considering these points:
1. Tools cannot be used by two machines at the same time.
2. If the same tool is required concurrently, additional tool sets are needed.
3. Try to minimize tool sets.

Format:
Tool Sets Needed
Sharing / Transitions
T1 <number> Set1: (...) Set2: (...)
T2 <number> Set1: (...) Set2: (...)

Schedule:
{schedule}

Tool Usage:
{tool_usage}
"""
)

# --- RAG + Context Prompt ---
rag_template = ChatPromptTemplate.from_template(
    """You are an expert in Job Shop Scheduling with integrated tool resource management.
Use this background knowledge:
{context}

Determine the tool sharing plan and minimum number of tool sets required.
Avoid overlaps and minimize total sets.

Format:
Tool Sets Needed
Sharing / Transitions
T1 <number> Set1: (...) Set2: (...)
T2 <number> Set1: (...) Set2: (...)

Schedule:
{schedule}

Tool Usage:
{tool_usage}
"""
)

# ============================================================
#  5. Run evaluation pipeline (2 samples only)
# ============================================================

df = pd.read_excel("/content/valid_50_with_tools_v3.xlsx")
results = []

for i, row in enumerate(df.head(2).iterrows(), start=1):
    _, row = row
    qid = row["ID"]
    schedule = row["Question"]
    tool_usage = row["Tool_Assignment"]

    print(f" Processing {qid}...")

    # --- Baseline ---
    base_msgs = baseline_template.format_messages(schedule=schedule, tool_usage=tool_usage)
    base_out = llm.invoke(base_msgs).content

    # --- Detailed ---
    det_msgs = detailed_template.format_messages(schedule=schedule, tool_usage=tool_usage)
    det_out = llm.invoke(det_msgs).content

    # --- RAG + Context ---
    docs = rag_db.similarity_search("tool sharing rules", k=2)
    context = "\n".join([d.page_content for d in docs])
    rag_msgs = rag_template.format_messages(
        context=context, schedule=schedule, tool_usage=tool_usage
    )
    rag_out = llm.invoke(rag_msgs).content

    print(f" Completed {qid}")

    results.append(
        {
            "ID": qid,
            "Schedule": schedule,
            "Tool_Usage": tool_usage,
            "Baseline_Output": base_out.strip(),
            "Detailed_Output": det_out.strip(),
            "RAG_Output": rag_out.strip(),
        }
    )

    time.sleep(0.5)

# ============================================================
#  6. Save final results
# ============================================================
os.makedirs("outputs", exist_ok=True)
out_path = "outputs/tool_sharing_comparison.xlsx"
pd.DataFrame(results).to_excel(out_path, index=False)
print(f"\n Results saved to {out_path}")
```
